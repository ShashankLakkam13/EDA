```python
#series is like dictionary 

#series is like pair

#series is like key value pair

#series can be converted into dictionaries and we can apply dictionary methods
```

create categoeical list


create numerical list

idea:
=======

convert series to dictionary


iterate through loop 

apply the condition 

and get different list for categorical and numerical data



```python
categorical=[]
numerical=[]
dictdata=dict(visadata.dtypes)

for i,j in dictdata.items():
    if j=="object":
        categorical.append(i)
    else:
        numerical.append(i)
print(categorical)
print(numerical)
```
